```python
import matplotlib.pyplot as pit
import numpy as np
from sklearn import datasets, linear_model, model_selection

X,y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
# Extract thr column at index 2 and reshape to a 2D array


X = X[:, 2]
print(X.shape)

X= X.reshape((-1, 1))
print(X.shape)
```

```python
# Split the data into training nad testing data. usually the test data is 1/3 of all the data ( i saw somewhere that it is 80/20. 80% percent training data 20% test data)

X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size=0.33)
```

```python
# create a linear regression model and train with out data

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# test it with our test data

y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='black')

plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')

plt.show()
```
